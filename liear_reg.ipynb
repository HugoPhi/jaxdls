```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import jax.numpy as jnp
from jax import grad, jit, vmap
from jax import random

key = random.PRNGKey(0)  # 使用任意整数作为种子
key, subkey = random.split(key)
```

```python
dim = (5, 3)

X = random.normal(subkey, (10000, 5))
w0 = random.normal(subkey, dim)
b0 = random.normal(subkey, 1)
print(f'w0 is: \n{w0}')
print(f'b0 is: {b0}')
```

```python
y = X @ w0 + b0
y += random.normal(subkey, (10000, y.shape[1])) * 0.01

print(y.shape)
```

```python
lo = lambda y, y_pred: jnp.mean((y - y_pred)**2)
loss = lambda w, b: lo(y, X @ w + b)
loss = jit(loss)  # time: 3.3542935848236084 -> 1.253995656967163 
```

```python
w, b = random.normal(subkey, dim), random.normal(subkey, 1)

alpha = 0.1
ll = []
import time
start = time.time()
for _ in range(100):
    ll.append(loss(w, b))
    dw = -grad(loss, argnums=0)(w, b)
    db = -grad(loss, argnums=1)(w, b)
    w += alpha * dw
    b += alpha * db

print(f'[*] time cost while training: {time.time() - start}')

print(f'[*] w after training: \n{w}')
print(f'[*] b after training: {b}')
```

```python
import matplotlib.pyplot as plt

plt.plot(range(len(ll)), ll, label='loss with epochs')
plt.legend()
plt.show()
```
