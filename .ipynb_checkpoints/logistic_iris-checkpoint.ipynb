```python
import jax.numpy as jnp
from jax import grad, vmap, jit
from jax import random

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import json

key = random.PRNGKey(0)

iris = sns.load_dataset('iris')

print(iris.head())
print()
print(f'[*] shape: {iris.shape}')
catg = np.unique(iris['species'].to_numpy())
print(f'[*] catagories: {catg}')
```

```python
name2num = {v: k for k, v in enumerate(catg)}
print(json.dumps(name2num, indent=4, ensure_ascii=False))

iris['species'] = iris['species'].map(name2num)
print()
print(iris.head())
```

```python
data = jnp.array(iris)
print(data[:10])
X = data[:, :-1]
y = data[:, -1].astype(jnp.int32)
print(X.shape, y.shape)
```

```python
indices = random.permutation(key, X.shape[0])
X = X[indices]
y = y[indices]


def one_hot(x: jnp.ndarray, num_class=3):
    if x.dtype != jnp.int32:
        raise ValueError('input array should be type jax.numpy.int32')

    res = jnp.zeros((x.shape[0], num_class))
    res = res.at[jnp.arange(x.shape[0]), x].set(1)

    return res

y_one_hot = one_hot(y)
print(y_one_hot[:10])
```

```python
split_rt = 0.7
NUM_TRAIN = int(X.shape[0] * split_rt)
X_train, X_test = X[:NUM_TRAIN], X[NUM_TRAIN:]
y_train, y_test = y_one_hot[:NUM_TRAIN], y_one_hot[NUM_TRAIN:]
```

```python
softmax = lambda x: jnp.exp(x) / jnp.sum(jnp.exp(x), axis=1, keepdims=True)  # keep dims here
predict_proba = lambda x, w, b: softmax(x @ w + b)
```

```python
w = random.normal(key, (X.shape[1], y_one_hot.shape[1]))
b = random.normal(key, 1)

loss = lambda w, b: jnp.sum(- y_train * jnp.log(predict_proba(X_train, w, b)), axis=1).mean()
loss = jit(loss)  # 128.762127161026 s -> 15.178180694580078 s

tloss = lambda w, b: jnp.sum(- y_test * jnp.log(predict_proba(X_test, w, b)), axis=1).mean()
```

```python
alpha = 0.01
epochs = 10000

lloss = []
lacc = []
tlloss = []
tlacc = []
import time
start = time.time()
for _ in range(epochs):
    lloss.append(loss(w, b))
    tlloss.append(tloss(w, b))
    
    dw = -alpha * grad(loss, argnums=0)(w, b)
    db = -alpha * grad(loss, argnums=1)(w, b)
    w += dw
    b += db
    
    y_pred = jnp.argmax(predict_proba(X_train, w, b), axis=1)
    y_true = jnp.argmax(y_train, axis=1)
    lacc.append(jnp.mean(y_true == y_pred))
    
    y_pred = jnp.argmax(predict_proba(X_test, w, b), axis=1)
    y_true = jnp.argmax(y_test, axis=1)
    tlacc.append(jnp.mean(y_true == y_pred))

time_cost = time.time() - start
print(f'training cost: {time_cost} s')
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy', color=color)
ax1.plot(range(epochs), lacc, color=color, label='Train Accuracy', linestyle='-')
ax1.plot(range(epochs), tlacc, color=color, label='Test Accuracy', linestyle='--')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  

color = 'tab:blue'
ax2.set_ylabel('Loss', color=color)  
ax2.plot(range(epochs), lloss, color=color, label='Train Loss', linestyle='-')
ax2.plot(range(epochs), tlloss, color=color, label='Test Loss', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color)

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles1 + handles2, labels1 + labels2, loc='lower right')

plt.title('Training and Testing Accuracy and Loss over Epochs')
fig.tight_layout()
plt.show()

print(f'final train, test acc : {lacc[-1]}, {tlacc[-1]}')
print(f'final train, test loss: {lloss[-1]}, {tlloss[-1]}')
```
